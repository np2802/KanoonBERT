```python
from google.colab import drive
import os
import random
import shutil
```

```python
# Mount Google Drive
drive.mount('/content/drive')
```

```python
# Define the path to your folders
folders = ['/content/drive/MyDrive/FYP/preprocessed_data/1950-1980',
           '/content/drive/MyDrive/FYP/preprocessed_data/1981-1996',
           '/content/drive/MyDrive/FYP/preprocessed_data/1997-2000',
           '/content/drive/MyDrive/FYP/preprocessed_data/2001-2024']
```

```python
# Function to select random files from each folder
def select_stratified_random_files(folders, total_files=500):
    num_folders = len(folders)
    files_per_folder = total_files // num_folders
    remainder = total_files % num_folders

    selected_files = []
    for i, folder in enumerate(folders):
        files = os.listdir(folder)
        num_files = files_per_folder + (1 if i < remainder else 0)
        selected_files.extend(random.sample(files, min(num_files, len(files))))
    return selected_files

# Create a new folder to copy selected files
output_folder = '/content/drive/MyDrive/FYP/Dataset/NewSelectedFiles'
os.makedirs(output_folder, exist_ok=True)
print("Folder created")

# Select random files stratified across all folders
random_files = select_stratified_random_files(folders, total_files=500)
print("Random files selected")

# Copy selected files to the new folder
for file_name in random_files:
    print(file_name)
    for folder in folders:
        source = os.path.join(folder, file_name)
        if os.path.exists(source):
            destination = os.path.join(output_folder, file_name)
            shutil.copy(source, destination)
            break
print("files selected copied")

# Write the list of file names to a new .txt file
output_file = '/content/drive/MyDrive/FYP/Dataset/NewSelectedFilesList.txt'
with open(output_file, 'w') as f:
    for file_name in random_files:
        f.write(file_name + '\n')

print(f"Total files selected: {len(random_files)}")
print(f"List of selected file names saved to {output_file}")
print(f"Selected files copied to {output_folder}")
```
